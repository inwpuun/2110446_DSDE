1. Load "ModifiedEdibleMushroom.csv" data from the link below (note: this data set has been preliminarily prepared.).

```python
import numpy as np
import pandas as pd
df = pd.read_csv('https://github.com/pvateekul/2110446_DSDE_2023s2/raw/main/code/Week03_ML/mushroom2020_dataset.csv')
```

```python
df['gill-size'].isnull().sum()
```

2. Drop rows where the target (label) variable is missing.



```python
df2=df.dropna(subset=['label'])
```

3. Drop the following variables:

'id','gill-attachment', 'gill-spacing', 'gill-size','gill-color-rate',

'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring-rate','stalk-color-below-ring-rate','veil-color-rate','veil-type'

```python

df2=df2.drop(['id','gill-attachment', 'gill-spacing', 'gill-size','gill-color-rate',
'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring-rate','stalk-color-below-ring-rate','veil-color-rate','veil-type'], axis='columns')
df2
```

4. Examine the number of rows, the number of digits, and whether any are missing.

```python
df2.info
```

5. Fill missing values by adding the mean for numeric variables and the mode for nominal variables.

```python
from sklearn.impute import SimpleImputer
isnull_num=df2.isnull().sum()
print(isnull_num)
modea=['cap-surface','bruises','odor','stalk-shape','ring-number','ring-type','spore-print-color','population','habitat']
meana=['cap-color-rate']
mode_impute=df2[modea]
mean_impute=df2[meana]
num_imp_mean=SimpleImputer(missing_values=np.NaN, strategy='mean')
# print(df2["cap-color-rate"].mean())
df2[meana]=num_imp_mean.fit_transform(mean_impute)


num_imp_mode=SimpleImputer(missing_values=np.NaN, strategy='most_frequent')
df2[modea]=num_imp_mode.fit_transform(mode_impute)

isnull_num=df2.isnull().sum()
print(isnull_num)
df2
```

6. Convert the label variable e (edible) to 1 and p (poisonous) to 0 and check the quantity. class0: class1

```python
mapper = {'e': 1, 'p': 0}
df2['label']= df2['label'].map(mapper)

count_1 = len(df2[df2['label'] == 1])
count_0 = len(df2[df2['label'] == 0])
print(count_0,count_1)
```

7. Convert the nominal variable to numeric using a dummy code with drop_first = True.

```python
subset_columns = ["cap-shape", "cap-surface","bruises","odor","stalk-shape","ring-number","ring-type","spore-print-color","population","habitat"]
dummy_df = pd.get_dummies(df2[subset_columns], drop_first=True)
df3 = pd.concat([df2, dummy_df], axis=1)
df3.drop(columns=subset_columns,inplace=True)
df3
```

```python
from sklearn.model_selection import train_test_split
y = df3.pop('label')
X = df3

X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=y,test_size=0.2, random_state=2020)
print(len(X_train),len(X_test))
X_train
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

rf_classifier = RandomForestClassifier(n_estimators=100, max_depth=6, min_samples_leaf=2,criterion='gini')

rf_classifier.fit(X_train, y_train)

# feature_importances = rf_classifier.feature_importances_

predictions = rf_classifier.predict(X_test)

accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)


conf_matrix = confusion_matrix(y_test, predictions)
print("Confusion Matrix:")
print(conf_matrix)


class_report = classification_report(y_test, predictions,digits=4)
print("Classification Report:")
print(class_report)
```

```python


param_grid1 = {
  'criterion':['gini','entropy'],
  'max_depth': [2,3,6],
  'min_samples_leaf':[2,5,10],
  'n_estimators': [100, 200],
  'random_state': [2020]
}

grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param_grid1, cv=5)
grid_search.fit(X_train, y_train)

print("Best Parameters:", grid_search.best_params_)


best_rf_model = grid_search.best_estimator_
best_rf_model.fit(X_train, y_train)


# import pickle
# filename = 'model.sav'
# pickle.dump(grid_search, open(filename, 'wb'))


# grid_search = pickle.load(open(filename, 'rb'))
# grid_search

y_pred = best_rf_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy on Test Set:", accuracy)

```

10. Predict the testing data set with confusion_matrix and classification_report.

```python
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)


class_report = classification_report(y_test, y_pred,digits=4)
print("Classification Report:")
print(class_report)
```
